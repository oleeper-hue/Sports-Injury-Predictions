```python
#get full list of nba players from NBA API, filtered for players with IDs that are relevant, and put their names and IDs in a DataFrame
#these IDs identify players for the NBA API
#relevant players played in the regular seasons 2017-18 to 2024-25
athletes = players.get_players()
ls = [[athlete['id'], athlete['full_name']] for athlete in athletes if athlete['id'] in rel_ids]
print(len(ls))
```

```python
nba_api = pd.DataFrame(ls, columns=['id', 'full_name'])
nba_api.head()
```

```python
#load DataFrame of information saved from the sportsdataverse/hoopR repository
path = 'C:\\Users\\oscar\\OneDrive\\Desktop\\HON 492\\espn_athlete_ids.csv'
hoopR = pd.read_csv(path)
hoopR.head()
```

```python
#check for athletes that appear twice in nba_api (the DataFrame) ... there are none
def check_dupes(ls):
    running_tally = []
    dups = []

    for athlete in ls:
        if athlete in running_tally:
            dups.append(athlete)
        running_tally.append(athlete)
    return dups

print(check_dupes(nba_api['full_name']))
```

```python
#check for athletes that appear twice in espn (the DataFrame)
print(check_dupes(hoopR['athlete_display_name']))
```

```python
#find the indices that have the duplicated athletes in hoopR DataFrame
dupes = check_dupes(hoopR['athlete_display_name'])
i = 0
dup_indices = []
for athlete in hoopR['athlete_display_name']:
    if athlete in dupes:
        dup_indices.append(i)
    i += 1

print(dup_indices)
print('number of duped athletes: ', len(dupes))
print('number of indices including duped athletes: ', len(dup_indices), '(this number should be double or more...)')
```

```python
#this just makes it easy to visualize
hoopR.iloc[dup_indices, 1].sort_values()
```

```python
#separate the records in hoopR into two DataFrames so that athletes with multiple IDs can be handled separately
hoopR_no_dupes = hoopR.drop(dup_indices)
hoopR_no_dupes.reset_index(drop=True, inplace=True)
hoopR_dupes = hoopR.loc[dup_indices]
hoopR_dupes.reset_index(drop=True, inplace=True)
```

```python
#check that 20 records are in the dupes DataFrame and the other 1316 are in the no_dupes DataFrame
print('shape of hoopR: ', hoopR.shape)
print('shape of hoopR_no_dupes: ', hoopR_no_dupes.shape)
print('shape of hoopR_dupes: ', hoopR_dupes.shape)
```

```python
#check if the athletes that are duplicated in hoopR are represented in the nba_api DataFrame ... 9/10 are represented.
dupes_in_nba_api = []
for athlete in hoopR_dupes['athlete_display_name'].unique():
    if athlete in nba_api['full_name'].values:
        dupes_in_nba_api.append(athlete)

print(dupes_in_nba_api)
```

```python
#display the name of the athlete not represented in the nba_api Dataframe and duplicated in the hoopR DataFrame
for athlete in hoopR_dupes['athlete_display_name'].unique():
    if athlete not in dupes_in_nba_api:
        print(athlete)
```

```python
#find the indices of the athletes duplicated in the hoopR DataFrame and separate the nba_api DataFrame by these athletes
i = 0
dup_indices = []
for name in nba_api['full_name']:
    if name in hoopR_dupes['athlete_display_name'].unique():
        dup_indices.append(i)
    i += 1
print(dup_indices)

nba_api_no_dupes = nba_api.drop(dup_indices)
nba_api_no_dupes.reset_index(drop=True, inplace=True)
nba_api_dupes = nba_api.loc[dup_indices]
nba_api_dupes.reset_index(drop=True, inplace=True)

#should be a 9 record difference in number of rows between the two new DataFrames
print('shape of nba_api: ', nba_api.shape)
print('shape of nba_api_no_dupes: ', nba_api_no_dupes.shape)
print('shape of nba_api_dupes: ', nba_api_dupes.shape)
```

```python
#lets find the common names in both and examine the names that are not matches
name_matches = []
for name in nba_api_no_dupes['full_name']:
    if name in hoopR_no_dupes['athlete_display_name'].values:
        name_matches.append(name)

print('shape of nba_api_no_dupes: ', nba_api_no_dupes.shape)
print('shape of hoopR_no_dupes: ', hoopR_no_dupes.shape)
print('number of matches: ', len(name_matches))
```
