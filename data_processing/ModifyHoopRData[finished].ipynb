**THE FOLLOWING ARE PROCEDURES TO CLEAN THE DATA IDENTIFIED IN createIDmarriagetable.ipynb**

```python
"""
Analysis in investigate_id_duplicates.ipynb I found that these changes needed to be made in the hoopR data so that the names are consistent 
across all records for players and under one ID and that the name matches the name in nba_api DataFrame...
    a. Jimmy Butler III to Jimmy Butler
    b. Reggie Bullock to Reggie Bullock Jr.
    c. Rondae Hollis Jefferson to Rondae Hollis-Jefferson
    d. O.G. Anunoby to OG Anunoby
    e. Omari Rasulala Spellman to Omari Spellman

I analyzed the athletes that have two different IDs in the hoopR data in investigate_name_duplicates.ipynb and found that these changes must be
made in the data and these IDs be chosen over the alternative ID...

DROP RECORDS

1. for Corey Brewer, records with player_ID 4415554 should be dropped from the hoopR data. Final ID for Corey Brewer: 3191.

2. for Daryl Macon, records with player_ID 4610145 should be dropped from the hoopR data. Final ID for Daryl Macon: 4066243.

3. for Henry Ellenson, records with the player_ID 4417428 should be dropped from the hoopR data. Final ID for Henry Ellenson: 3906522.

4. for John Jenkins, records with the player_ID 4415789 should be dropped from the hoopR data. Final ID for John Jenkins: 6594.

5. for Mitchell Creek, records with the player_ID 4412474 should be dropped from the hoopR data. Final ID for Mitchell Creek: 4236300.

6. for Isaiah  Canaan, records with the player_ID 4412182 should be dropped from the hoopR data. Final ID for Isaiah Canaan: 2490589.

MODIFY RECORDS

7. for Greg Monroe, the one record with the player_ID 4412680 should be changed to 4260. Final ID for Greg Monroe: 4260.

8. for Ray Spalding, the one record with the player_ID 4415522 should be changed to 3908806. Final ID for Ray Spalding: 3908806.

9. for Wayne Selden, all records with the player_ID 4895499 should be changed to 3059316. Final ID for Wayne Selden: 3059316.
"""
```

```python
#modify records: player names
rename = [
    ['Jimmy Butler III', 'Jimmy Butler'],
    ['Reggie Bullock', 'Reggie Bullock Jr.'],
    ['Rondae Hollis Jefferson', 'Rondae Hollis-Jefferson'],
    ['O.G. Anunoby', 'OG Anunoby'],
    ['Omari Rasulala Spellman', 'Omari Spellman']
]
#delete records: player IDs
ID_delete_rows = [
    4415554,
    4610145,
    4417428,
    4415789,
    4412474,
    4412182
]
#modify records: player IDs
reID = [
    [4412680, 4260],
    [4415522, 3908806],
    [4895499, 3059316]
]
```

```python
#modify records: player names

#def function that counts all instances of rows containing a list of values in a column in a DataFrame
def val_count(values, col, DataFrame):
    count = 0
    for val in values:
        rows, cols = DataFrame[DataFrame[col] == val].shape
        count += rows
    return count
#count all rows containing the player names that will be changed
old_names = [rename[0][0], rename[1][0], rename[2][0], rename[3][0], rename[4][0]]
print('Rows containing the old player names: ', val_count(old_names, 'athlete_display_name', df))

#change the names
for old, new in rename:
    df['athlete_display_name'] = df['athlete_display_name'].replace(old, new)
print('-----OLD PLAYER NAMES CHANGED-----')

#count all rows containing the player names to show they were changed
print('Rows containing the old player names: ', val_count(old_names, 'athlete_display_name', df))
```

```python
#delete records: player IDs

#count all rows containing the player IDs that will be deleted
print('Rows containing the ID that must be deleted: ', val_count(ID_delete_rows, 'athlete_id', df))

#delete the player IDs
for idx in ID_delete_rows:
    df = df[df['athlete_id'] != idx]
print('-----PLAYER IDS DELETED-----')

#count all rows containing the player IDs that were deleted
print('Rows containing the ID that must be deleted: ', val_count(ID_delete_rows, 'athlete_id', df))
```

```python
#same process above for changed player names but for IDs

#check for row count of rows containing the IDs that must be modified
old_IDs = [reID[0][0], reID[1][0], reID[2][0]]
print('Rows containing the old player IDs: ', val_count(old_IDs, 'athlete_id', df))

#change the player IDs
for old, new in reID:
    df['athlete_id'] = df['athlete_id'].replace(old, new)
print('-----OLD PLAYER IDS CHANGED-----')

#check for row count of rows containing the IDs that were modified
old_IDs = [reID[0][0], reID[1][0], reID[2][0]]
print('Rows containing the old player IDs: ', val_count(old_IDs, 'athlete_id', df))
```
