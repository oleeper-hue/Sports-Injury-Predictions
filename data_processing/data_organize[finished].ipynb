```python
from datetime import datetime, date, timedelta

table_start = string_to_date('2017-09-17')
table_end = string_to_date('2025-04-13')

def string_to_date(date_str):
    return datetime.strptime(date_str, "%Y-%m-%d")

def date_to_string(date):
    return date.strftime('%Y-%m-%d')

def player_last_game(player_key):
    player_games = game_fact.loc[game_fact['player_key'] == player_key, :]
    player_games = player_games.sort_values(by='game_date')
    date_str = player_games.tail(1)['game_date'].values[0]
    return string_to_date(date_str)

def player_draft_date(player_key):
    #date_str = player_dim.loc[player_dim['player_key'] == player_key, 'draft_date'].values[0]
    date_str = str(player_dim.loc[player_dim['player_key'] == player_key, 'from_year'].values[0]) + "-06-30"
    return string_to_date(date_str)

def player_birthdate(player_key):
    date_str = player_dim.loc[player_dim['player_key'] == player_key, 'birthdate'].values[0].split("T")[0]
    return string_to_date(date_str)

def player_window_start(player_key):
    draft_date = player_draft_date(player_key)
    if draft_date < table_start:
        return table_start
    else:
        return draft_date

def player_dates(player_key):
    window_end = player_last_game(player_key)
    window_start = player_window_start(player_key)

    days = []
    current_day = window_start
    while current_day != window_end:
        days.append(current_day)
        current_day += timedelta(days=1)
    days.append(window_end)
    return days

def team_on_player_date(player_key, date_str):
    player_games = game_fact.loc[game_fact['player_key'] == player_key, ['game_date', 'team_key']]
    return player_games.loc[player_games['game_date'] == date_str, 'team_key'].iloc[0]

def player_sparse_team_keys(player_key, player_dates):
    team_by_day = []
    
    for date in player_dates:
        date = date_to_string(date)
        if date in game_fact.loc[game_fact['player_key'] == player_key, 'game_date'].values:
            team_by_day.append(team_on_player_date(player_key, date))
        else:
            team_by_day.append(-1)
                
    return team_by_day

team_by_day = []

def populate_sparse(sparse_teams): #goes until it finds a non -1 key and fills all past ids and then resets
    current_stretch = []
    
    for i in range(len(sparse_teams)):
        current_stretch.append(i)
        if sparse_teams[i] != -1:
            for idx in current_stretch:
                sparse_teams[idx] = sparse_teams[i]
            current_stretch = []
            past_team_key = sparse_teams[i]
        elif i == len(sparse_teams) - 1: #fill up last few entries
            for idx in current_stretch:
                sparse_teams[idx] = past_team_key
        else:
            continue
    return sparse_teams

def difference_days(date2, date1): #difference between datetime objects in days
    difference = date2 - date1
    return difference.days

def all_days_age(player_key, player_dates):
    birthdate = player_birthdate(player_key)
    ages = []
    for date in player_dates:
        ages.append(difference_days(date, birthdate))

    return ages

def all_days_exp(player_key, player_dates):
    draft_date = player_draft_date(player_key)
    experience = []
    for date in player_dates:
        experience.append(difference_days(date, draft_date))

    return experience

def player_IL_dict(player_key):
    player_IL = IL_update_fact.loc[IL_update_fact['player_key'] == player_key, ['date', 'relinquished/acquired']].sort_values(by='date')
    dates = player_IL['date'].values
    ra = player_IL['relinquished/acquired'].values

    dates_to_ra = {}
    for i in range(len(dates)):
        dates_to_ra[dates[i]] = ra[i]

    if not dates_to_ra:
        return {'2017-09-01':'a'}
    else:
        return dates_to_ra

def prestatus(player_key):
    window_start = player_window_start(player_key) #draft day or (for older players) 2017-09-17
    IL_dict = player_IL_dict(player_key)
    update_dates = list(IL_dict.keys())
    oldest_IL_record = string_to_date(update_dates[0])

    if oldest_IL_record > window_start:
        return False

    update_dates = [string_to_date(date) for date in update_dates] #turn into datetime for comparison with current_day

    current_day = window_start
    while current_day not in update_dates:
        current_day = current_day - timedelta(days=1)

    current_day = date_to_string(current_day) #current_day to string to match the keys for the dict
    if IL_dict[current_day] == 'r':
        return True
    else:
        return False

def player_status_ls(player_key, player_dates):
    IL_dict = player_IL_dict(player_key)
    status = []
    current_status = prestatus(player_key)
    
    for date in player_dates:
        date = date_to_string(date)
        if date in list(IL_dict.keys()):
            if IL_dict[date] == 'r' and current_status == True:
                status.append(current_status)
                continue
            elif IL_dict[date] == 'r' and current_status == False:
                current_status = not current_status
                status.append(current_status)
            elif IL_dict[date] == 'a' and current_status == True:
                current_status = not current_status
                status.append(current_status)
            elif IL_dict[date] == 'a' and current_status == False:
                status.append(current_status)
                continue
        else:
            status.append(current_status)
    return status
        
    
    
#print(player_last_game(1626147))
#print(player_window_start(1626147))
#print(player_draft_date(1626147))
#print(player_birthdate(1626147))
#dates = player_dates(1626147)
#sparse = player_sparse_team_keys(1626147, dates)
#team_kys = populate_sparse(sparse)
#IL_status = player_status_ls(1626147, dates)
#age = all_days_age(1626147, dates)
#exp = all_days_exp(1626147, dates)
#print('REPORT')
#print('TEAM KEYS: ', len(team_kys))
#print('IL STATUS: ', len(IL_status))
#print('AGE: ', len(age))
#print('EXP: ', len(exp))
```
